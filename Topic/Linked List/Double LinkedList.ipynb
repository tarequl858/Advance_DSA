<h1>Traversal in Doubly Linked List</h1>

```python
class Node:
    def __init__ (self,data):
        self.data = data
        self.prev = None
        self.next = None
```

```python
def forward_traversal(head):
    forward_array = []
    curr = head
    while curr is not None:
        forward_array.append(curr.data)
        curr = curr.next
    backward_array = []
    backward_array = forward_array[::-1]
    return forward_array, backward_array
```

```python
if __name__ == "__main__":
    head = Node(50)
    head.next = Node(60)
    head.next.next = Node(20)
    head.next.next.next = Node(70)

    fwd, bwd = forward_traversal(head)
    print(fwd)
    print(bwd)
```

```python
def forward_traversal(head):
    forward_array = []
    curr = head
    while curr is not None:
        forward_array.append(curr.data)
        curr = curr.next
    backward_array = []
    while head is not None:
        backward_array.insert(0,head.data)
        head = head.next
    return forward_array, backward_array
```

```python
from collections import deque
def displayList(head):
    forward_array = []
    backward_array = deque()
    curr = head
    while curr is not None:
        forward_array.append(curr.data)
        backward_array.appendleft(curr.data)
        curr = curr.next
    return forward_array, list(backward_array)
```

```python
def forward_recursive(curr):
    if curr is not None:
        print(curr.data,end=" ")
        forward_traversal(curr.next)
```

```python
def backward_recursive(curr):
    if curr is not None:
        print(curr.data,end=" ")
        backward_traversal(curr.prev)
```

```python
if __name__ == "__main__":
    head = Node(50)
    head.next = Node(60)
    head.next.prev = head
    head.next.next = Node(20)
    head.next.next.prev = head.next
    head.next.next.next = Node(70)
    head.next.next.next.prev = head.next.next
    tail = head.next.next.next
    forward_recursive(head)
    print()
    backward_recursive(tail)
```

<h1>Insert a Node at Front</h1>

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        self.prev = None
```

```python
def insert_at_front(head,data):
    new_node = Node(data)
    new_node.next = head
    head.prev = new_node
    return new_node
```

```python
def insert_at_front_iterative_approach(curr):
    while curr is not None:
        print(curr.data,end=" ")
        if curr.next is not None:
            print("->",end="")
        curr = curr.next
```

```python
if __name__ == "__main__":
    new_data = int(input("Enter a new data: "))
    head = Node(50)
    head.next = Node(60)
    head.next.prev = head
    head.next.next = Node(20)
    head.next.next.prev = head.next
    head.next.next.next = Node(70)
    head.next.next.next.prev = head.next.next
    head = insert_at_front(head,new_data)
    insert_at_front_iterative_approach(head)
```
